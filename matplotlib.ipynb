```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt     
```

## Matplotlib Basics

```python
x = np.linspace(0,5,11)
x 
y = x ** 2
y 
```

```python
plt.plot(x,y)
plt.title('Graph')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')    
```

## Subplot in Matplotlib

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(y,x)
plt.subplot(2,2,3)
plt.plot(x **2,x **2)
plt.subplot(2,2,4)
plt.plot(x,x **0.5)
```

## Object Orinted Way

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
axes1.set_title('Graph 1') 
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes2.plot(y,x)
```
